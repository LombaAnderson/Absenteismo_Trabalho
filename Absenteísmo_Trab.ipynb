# Visualização de Dados

```python
plt.style.use('fivethirtyeight')
```

```python
# Verificando a distribuição dos dados 
plt.figure(1, figsize = (15, 6))
n = 0
for x in [ 'Idade', 'Num Filhos','Indisciplina']:
    n += 1
    plt.subplot(1, 3, n)
    plt.subplots_adjust(hspace = 0.5 , wspace = 0.5)
    sns.distplot(dados_trab[x], bins = 25)
    plt.title('{}'.format(x))
plt.show()
```

```python
# Verificando a distribuição concernente na idade, gastos com transporte e Faltas ao trabalho
plt.figure(1, figsize = (15, 6))
n = 0
for x in [ 'Idade', 'Gastos Transporte','Faltas Medicao em Horas']:
    n += 1
    plt.subplot(1, 3, n)
    plt.subplots_adjust(hspace = 0.5 , wspace = 0.5)
    sns.distplot(dados_trab[x], bins = 25)
    plt.title('{}'.format(x))
plt.show()
```

```python
# Verificando idade, tempo de serviço e indisciplina 
plt.figure(1, figsize = (15, 6))
n = 0
for x in [ 'Idade', 'Tempo Servico','Indisciplina']:
    n += 1
    plt.subplot(1, 3, n)
    plt.subplots_adjust(hspace = 0.5 , wspace = 0.5)
    sns.distplot(dados_trab[x], bins = 25)
    plt.title('{}'.format(x))
plt.show()
```
